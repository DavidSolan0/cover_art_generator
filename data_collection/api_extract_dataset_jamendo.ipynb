```python
import os
import requests
import pandas as pd
import config as cf

# Define URL
base_url = "https://api.jamendo.com/v3.0"

# Define API and ID
api_key = cf["jamendo_token"]
id_client = cf["jamendo_id"]

# Make folders
os.makedirs("songs", exist_ok=True)
os.makedirs("covers", exist_ok=True)


# obtain music and covers
def get_songs_with_covers_and_genre():
    # generate url
    search_url = f"{base_url}/tracks/?client_id={id_client}&limit=10&search='w'&include=musicinfo"

    # request api
    response = requests.get(search_url)
    results = response.json()

    # Create DF
    df = pd.DataFrame(columns=["ID", "Canción", "Portada de álbum", "Género principal"])

    # Loop over songs and get info
    for track in results["results"]:
        song_id = track["id"]
        song_name = track["name"]
        album_cover_url = track["album_image"]
        # primary_genre = track["tags"]["genres"][0]["name"]
        primary_genre = "Hip-Hop"
        # Download Song
        audio_url = track["audio"]
        audio_response = requests.get(audio_url)
        audio_file_path = f"songs/{song_id}.mp3"
        with open(audio_file_path, "wb") as audio_file:
            audio_file.write(audio_response.content)

        # Download Cover
        album_cover_response = requests.get(album_cover_url)
        album_cover_file_path = f"covers/{song_id}.jpg"
        with open(album_cover_file_path, "wb") as album_cover_file:
            album_cover_file.write(album_cover_response.content)

        # Agregate to DF
        df = df.append(
            {
                "ID": song_id,
                "Canción": song_name,
                "Portada de álbum": album_cover_file_path,
                "Género principal": primary_genre,
            },
            ignore_index=True,
        )

    # To CSV
    df.to_csv("canciones.csv", index=False)


# Run function
get_songs_with_covers_and_genre()
```
